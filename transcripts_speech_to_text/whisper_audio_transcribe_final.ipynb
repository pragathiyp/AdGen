```python
import os
import whisper

# Set the root folder where your dataset resides
root_folder = "/content/drive/MyDrive/dataset_all1"

# Define the subfolder containing the audio files
subfolder = "00001"

# Path to the audio files
audio_folder = os.path.join(root_folder, "audio", subfolder)

# Load the whisper model
model = whisper.load_model("large")  # You can change the model size if needed

# Function to transcribe audio files
def transcribe_audio_files(audio_folder, model):
    transcripts = {}
    audio_files = os.listdir(audio_folder)
    for audio_file in audio_files:
        if audio_file.endswith(".mp3"):
            audio_path = os.path.join(audio_folder, audio_file)
            result = model.transcribe(audio_path)
            transcripts[audio_file] = result
    return transcripts

# Get transcripts for audio files in the specified folder
transcripts = transcribe_audio_files(audio_folder, model)

# Print transcripts
for audio_file, transcript in transcripts.items():
    print(f"Transcript for {audio_file}: {transcript}")
```

```python
import os
import whisper

# Set the root folder where your dataset resides
root_folder = "/content/drive/MyDrive/dataset_all1/00002"

# Load the Whisper model with CUDA device 0
model = whisper.load_model("large-v2", device="cuda:0")  # You can change the model size if needed

# Function to transcribe audio files in the subfolders
def transcribe_audios_in_folder(folder_path):
    for item in os.listdir(folder_path):
        item_path = os.path.join(folder_path, item)
        if os.path.isdir(item_path):
            # Recursively traverse subfolders
            transcribe_audios_in_folder(item_path)
        elif item.endswith('.mp3'):  # Check if the item is an mp3 audio file
            audio_path = item_path
            print(f"Transcribing: {audio_path}")
            # Transcribe the audio file
            result = model.transcribe(audio_path)
            # Extract the transcript from the result
            transcript = result['text']
            # Create the transcript file path
            transcript_file_name = os.path.splitext(item)[0] + '_whisper_transcript.txt'
            transcript_file_path = os.path.join(os.path.dirname(audio_path), transcript_file_name)
            # Write the transcript to the file
            with open(transcript_file_path, 'w') as f:
                f.write(transcript)
            print(f"Transcript saved to: {transcript_file_path}")

# Start transcription process from the 00002 folder
transcribe_audios_in_folder(root_folder)
```
