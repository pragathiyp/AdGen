```python
import os
import requests
from PIL import Image
import torch
from transformers import AutoProcessor, Blip2ForConditionalGeneration

# Load the BLIP-2 model and processor
processor = AutoProcessor.from_pretrained("Salesforce/blip2-opt-2.7b")
model = Blip2ForConditionalGeneration.from_pretrained("Salesforce/blip2-opt-2.7b", torch_dtype=torch.float16)

# Use GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
```

```python

# Function to generate caption using BLIP-2 model
def generate_caption(image_path, max_length=4000):
    image = Image.open(image_path).convert("RGB")
    prompt = "this is a picture of"
    inputs = processor(image, text=prompt, return_tensors="pt").to(device, torch.float16)
    generated_ids = model.generate(**inputs, max_length=max_length)
    caption = processor.batch_decode(generated_ids, skip_special_tokens=True)[0].strip()
    return caption

# Function to process folders and images
def process_folders(root_folder):
    for root, dirs, files in os.walk(root_folder):
        for d in dirs:
            subfolder_path = os.path.join(root, d)
            if d.endswith("-images"):
                image_files = [os.path.join(subfolder_path, f) for f in os.listdir(subfolder_path) if f.lower().endswith((".jpg", ".jpeg", ".png"))]
            else:
                image_files = [os.path.join(root, d, f) for f in os.listdir(os.path.join(root, d)) if f.lower().endswith((".jpg", ".jpeg", ".png"))]

            for img in image_files:
                caption = generate_caption(img)
                print(caption)
                text_file_path = os.path.splitext(img)[0] + ".txt"
                with open(text_file_path, "w") as f:
                    f.write(caption)

# Example usage:
root_folder = "/content/drive/MyDrive/raw_topics_transcripts"
process_folders(root_folder)
```
